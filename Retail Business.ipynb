```python
print("Product Category Distribution:\n", df['Product_Category'].value_counts())
print("\nRegion Distribution:\n", df['Region'].value_counts())
print("\nCustomer Type Distribution:\n", df['Customer_Type'].value_counts())
```

Business Data Analysis

```python
region_revenue = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
region_revenue
```

```python
category_revenue = df.groupby('Product_Category')['Revenue'].sum().sort_values(ascending=False)
category_revenue
```

```python
customer_revenue = df.groupby('Customer_Type')['Revenue'].sum()
customer_revenue
```

```python
df['Month'] = df['Order_Date'].dt.month
monthly_sales = df.groupby('Month')['Revenue'].sum()
monthly_sales

```

```python
discount_impact = df.groupby('Discount_Percent')['Revenue'].mean()
discount_impact
```

Data Visualization

```python
category_revenue.plot(kind='bar')
plt.title("Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_sales.plot(marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
category_revenue.plot(kind='pie', autopct='%1.1f%%')
plt.title("Product Category Contribution")
plt.ylabel("")
plt.show()
```

```python
plt.hist(df['Revenue'], bins=25)
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df['Discount_Percent'], df['Revenue'])
plt.title("Discount vs Revenue")
plt.xlabel("Discount (%)")
plt.ylabel("Revenue")
plt.show()
```

```python
ig, axes = plt.subplots(1,2, figsize=(12,5))

# Region Revenue
region_revenue.plot(kind='bar', ax=axes[0])
axes[0].set_title("Region Revenue")

# Customer Type
customer_revenue.plot(kind='bar', ax=axes[1])
axes[1].set_title("Customer Revenue")

plt.tight_layout()
plt.show()
```

Key Insights

- Certain regions contribute significantly higher revenue
- Top product categories dominate overall sales
- Returning customers generate more revenue than new customers
- Moderate discounts improve sales, excessive discounts reduce profit margins
- Sales show seasonal trends across months

 Business Recommendations

- Focus marketing strategies on high-performing regions
- Optimize discount strategy to avoid revenue loss
- Retain customers using loyalty programs
- Increase stock for top-selling categories
